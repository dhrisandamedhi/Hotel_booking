#4.Data Preparation

Lets build boxplots to see Outilier better.

```python
#Check outlier 
columns = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies', 'required_car_parking_spaces', 'adr', 'previous_cancellations', 'previous_bookings_not_canceled', 'booking_changes']
n = 1
plt.figure(figsize=(20,15))

for column in columns:
  plt.subplot(4,4,n)
  n = n+1
  sns.boxplot(hotel[column])
  plt.tight_layout()
```

We remove the outliers by using conditionals to get and update these values.

```python
#Imputing the Outliers
hotel.loc[hotel.lead_time > 500, 'lead_time'] = 500
hotel.loc[hotel.stays_in_weekend_nights >=  5, 'stays_in_weekend_nights'] = 5
hotel.loc[hotel.adults > 4, 'adults'] = 4
hotel.loc[hotel.previous_bookings_not_canceled > 0, 'previous_bookings_not_canceled'] = 1
hotel.loc[hotel.previous_cancellations > 0, 'previous_cancellations'] = 1
hotel.loc[hotel.stays_in_week_nights > 10, 'stays_in_week_nights'] = 10
hotel.loc[hotel.booking_changes > 5, 'booking_changes'] = 5
hotel.loc[hotel.babies > 8, 'babies'] = 0
hotel.loc[hotel.required_car_parking_spaces > 5, 'required_car_parking_spaces'] = 0
hotel.loc[hotel.children > 8, 'children'] = 0
hotel.loc[hotel.adr > 1000, 'adr'] = 1000
```

```python
hotel.describe()
```

->We removed the outliers. Our data is clean now.

```python
#combine children and babies together as kids
hotel['Kids'] = hotel.children+ hotel.babies
#Combine total mumbers by adding kids and adults
hotel['total_members'] = hotel.Kids + hotel.adults
```

```python
#convert int to str
hotel['arrival_date_year']=hotel['arrival_date_year'].astype('str')
hotel['arrival_date_day_of_month']=hotel['arrival_date_day_of_month'].astype('str')
hotel['arrival_date_month']=hotel['arrival_date_month'].astype('str')

```

```python
#combine date,month,year 
hotel['arrival_date']=hotel['arrival_date_day_of_month']+'-'+hotel['arrival_date_month']+'-'+hotel['arrival_date_year']
hotel['arrival_date'] = pd.to_datetime(hotel['arrival_date'], errors='coerce')
```

```python
# Adding total staying days in hotels
hotel['total_stay'] = hotel['stays_in_weekend_nights']+hotel['stays_in_week_nights']
```
