#**4.Data Preparation**

Lets build boxplots to see Outilier better.

```python
#Check outlier 
columns = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies', 'required_car_parking_spaces', 'adr', 'previous_cancellations', 'previous_bookings_not_canceled', 'booking_changes']
n = 1
plt.figure(figsize=(20,15))

for column in columns:
  plt.subplot(4,4,n)
  n = n+1
  sns.boxplot(hotel[column])
  plt.tight_layout()
```

->We remove the outliers by using conditionals to get and update these values.

```python
#Imputing the Outliers
hotel.loc[hotel.lead_time > 500, 'lead_time'] = 500
hotel.loc[hotel.stays_in_weekend_nights >=  5, 'stays_in_weekend_nights'] = 5
hotel.loc[hotel.adults > 4, 'adults'] = 4
hotel.loc[hotel.previous_bookings_not_canceled > 0, 'previous_bookings_not_canceled'] = 1
hotel.loc[hotel.previous_cancellations > 0, 'previous_cancellations'] = 1
hotel.loc[hotel.stays_in_week_nights > 10, 'stays_in_week_nights'] = 10
hotel.loc[hotel.booking_changes > 5, 'booking_changes'] = 5
hotel.loc[hotel.babies > 8, 'babies'] = 0
hotel.loc[hotel.required_car_parking_spaces > 5, 'required_car_parking_spaces'] = 0
hotel.loc[hotel.children > 8, 'children'] = 0
hotel.loc[hotel.adr > 1000, 'adr'] = 1000
```

```python
hotel.describe()
```

->We removed the outliers. Our data is clean now.

```python
#combine children and babies together as kids
hotel['Kids'] = hotel.children+ hotel.babies
#Combine total mumbers by adding kids and adults
hotel['total_members'] = hotel.Kids + hotel.adults
```

```python
#convert int to str
hotel['arrival_date_year']=hotel['arrival_date_year'].astype('str')
hotel['arrival_date_day_of_month']=hotel['arrival_date_day_of_month'].astype('str')
hotel['arrival_date_month']=hotel['arrival_date_month'].astype('str')

```

```python
#combine date,month,year 
hotel['arrival_date']=hotel['arrival_date_day_of_month']+'-'+hotel['arrival_date_month']+'-'+hotel['arrival_date_year']
hotel['arrival_date'] = pd.to_datetime(hotel['arrival_date'], errors='coerce')
```

```python
# Adding total staying days in hotels
hotel['total_stay'] = hotel['stays_in_weekend_nights']+hotel['stays_in_week_nights']
```

Q.Which hotel has higher bookings cancellation rate.

```python
# Selecting and counting number of cancelled bookings for each hotel.
cancelled_data = hotel[hotel['is_canceled'] == 1]
cancel_grp = cancelled_data.groupby('hotel')
D1 = pd.DataFrame(cancel_grp.size()).rename(columns = {0:'total_cancelled_bookings'})

# Counting total number of bookings for each type of hotel
grouped_by_hotel = hotel.groupby('hotel')
total_booking = grouped_by_hotel.size()
D2 = pd.DataFrame(total_booking).rename(columns = {0: 'total_bookings'})
D3 = pd.concat([D1,D2], axis = 1)

# Calculating cancel percentage 
D3['cancel_%'] = round((D3['total_cancelled_bookings']/D3['total_bookings'])*100,2)
D3
```

```python
plt.figure(figsize = (5,4))
sns.barplot(x = D3.index, y = D3['cancel_%'])
plt.title("'Cancelation rates in City hotel and Resort hotel",fontweight="bold", size=12)
plt.show()
```

Let us divide our customers in three categories of single, couple and family/friends. then check their booking %

```python
## Select single, couple, multiple adults and family
single   = not_canceled[(not_canceled['adults']==1) & (not_canceled['children']==0) & (not_canceled['babies']==0)]
couple   = not_canceled[(not_canceled['adults']==2) & (not_canceled['children']==0) & (not_canceled['babies']==0)]
family   = not_canceled[not_canceled['adults'] + not_canceled['children'] + not_canceled['babies'] > 2]

reindex = ['January', 'February','March','April','May','June','July','August','September','October','November','December']

fig, ax = plt.subplots(figsize=(12, 8))

for type in ['single', 'couple', 'family']:
  hotel = eval(type).groupby(['arrival_date_month']).size().reset_index().rename(columns = {0:'arrival_num'})
  hotel['arrival_date_month'] = pd.Categorical(hotel['arrival_date_month'],categories=reindex,ordered=True)
  sns.lineplot(data=hotel, x= 'arrival_date_month', y='arrival_num', label=type, ax=ax)
  
plt.grid()
plt.show()
```
